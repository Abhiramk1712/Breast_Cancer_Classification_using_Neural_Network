# Breast Cancer Classification using Neural Network

Breast Cancer is classified into two:- 

    1. Benign tumor
    2. Malignant tumor

1. Benign Tumor

- Non-cancerous

- Capsulated

- Slow growing

- Do not spread to other parts of the body

- Cells are normal

2. Malignant Tumor

- Cancerous

- Non-capsulated

- Fast growing

- Spreads to other parts of the body

- Cells have large, dark nuclei, may have abnormal growth

### About dataset

- Fine needle aspiration is a type of biopsy procedure.

- In fine needle aspiration, a thin needle is inserted into an area of abnormal appearing tissue or body fluid. As with other types of biopsies, the sample collected during this procedure can help make a diagnosis or rule out conditions such as cancer.

### Import the dependencies 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

### Separating the features and target 

```python
X = df.drop(columns = 'label', axis = 1)
Y = df['label']
```

```python
print(X)
```

```python
print(Y)
```

### Spliting the data into training & testing datasets

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

[test_size = 0.2] - 20% of data will be used as testing data
[random_state=2] - to get reproducible results whenever we run our code

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Standardizing the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

### Building the Neural Network

```python
# importing Tensorflow and Keras (ML & DL libraries)
import tensorflow as tf
tf.random.set_seed(3) # everytime we run the 
# code random weights and inputs are 
# initialized and the accuracy score changes, 
# in order to avoid that we set the raondom seed 
# to a fixed state. Everytime we run the code we 
# will get the same accuracy score.
from tensorflow import keras
```

```python
# setting up the layers of neural network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #input layer (matrix input is converted into single dimensional layer)
    keras.layers.Dense(20, activation='relu'), #hidden layer
    keras.layers.Dense(2, activation='sigmoid') #output layer  
])
```

```python
# Compiling the neural network
model.compile(optimizer = 'adam',
             loss = 'sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

#### Loss function is the difference between true value and predicted 

#### Loss function is inversely proportional to accuracy score

```python
# Training the neural network
history = model.fit(X_train_std, Y_train, validation_split = 0.1, epochs = 10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'], loc = 'upper right')
```

```python
# Accuracy of the model on test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

### Building the predictive system 

```python
input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)

# change the input_data to a numpy array
input_data_as_np_arr = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshape = input_data_as_np_arr.reshape(1,-1)

# Standardizing the input_data
input_data_std = scaler.transform(input_data_reshape)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
    print("The tumor is Malignant.")
else:
    print("The tumor is Benign.")
```
